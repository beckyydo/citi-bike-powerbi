```python
# Import Dependencies
import pandas as pd
```

```python
# Import monthly datasets
citi_01 = pd.read_csv("asset/data/201911-citibike-tripdata.csv")
citi_02 = pd.read_csv("asset/data/201912-citibike-tripdata.csv")
citi_03 = pd.read_csv("asset/data/202001-citibike-tripdata.csv")
citi_04 = pd.read_csv("asset/data/202002-citibike-tripdata.csv")
```

```python
# Concat the multiple datasets into one dataset
citi_df = pd.concat([citi_01, citi_02, citi_03, citi_04], ignore_index = True)
```

```python
# Display preview of dataframe
citi_df.head()
```

```python
# Export final dataset for dashboard
citi_df.to_csv("asset/data/citi_df.csv")
```
